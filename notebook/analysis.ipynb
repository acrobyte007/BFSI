```python
import pandas as pd
import re

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

df = df.drop_duplicates(subset=["url"])

df["combined_text"] = df["title"].fillna('') + " " + df["opening_text"].fillna('')
df = df.drop_duplicates(subset=["combined_text"])

df["date"] = pd.to_datetime(df["date"], errors="coerce")

df["source_name"] = df["source_name"].astype(str).str.strip().str.lower()

if "sentiment" in df.columns:
    df["sentiment"] = df["sentiment"].astype(str).str.strip().str.capitalize()

if "reach" in df.columns:
    df["reach"] = pd.to_numeric(df["reach"], errors="coerce")

def clean_text(text):
    if pd.isna(text):
        return ""
    text = text.lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-zA-Z0-9₹$\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df["clean_text"] = df["combined_text"].apply(clean_text)

df["processed_text"] = df["clean_text"]

final_columns = [
    "date",
    "url",
    "source_name",
    "title",
    "opening_text",
    "hit_sentence",
    "processed_text",
    "sentiment",
    "reach"
]

final_columns = [col for col in final_columns if col in df.columns]

final_df = df[final_columns]

final_df.to_csv("E:\BFSI\BFSI\data\cleaned_data.csv", index=False)

print("Data cleaning completed. File saved as cleaned_data.csv")
```
